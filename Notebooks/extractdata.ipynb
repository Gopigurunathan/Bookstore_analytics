```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine

# Database connection details
DB_HOST = "localhost"
DB_PORT = "3306"  # Default MySQL port
DB_NAME = "gravity_books"
DB_USER = "root"
DB_PASSWORD = "Gopi2708" 

# Function to fetch table data as a DataFrame
def fetch_table_as_dataframe(table_name, conn):
    query = f"SELECT * FROM {table_name};"
    try:
        df = pd.read_sql(query, conn)
        return df
    except Exception as e:
        print(f"Error fetching data from {table_name}: {e}")
        return None

# List of tables to fetch
tables = [
    "author", "publisher", "book_language", "book", "book_author",
    "address_status", "country", "address", "customer", "customer_address",
    "shipping_method", "cust_order", "order_status", "order_line", "order_history"
]

# Dictionary to store DataFrames
dataframes = {}

# Establish a database connection
try:
    # Create an SQLAlchemy engine
    engine = create_engine(f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}")
    
    with engine.connect() as connection:
        # Fetch each table as a DataFrame
        for table in tables:
            print(f"Fetching data for table: {table}")
            dataframes[table] = fetch_table_as_dataframe(table, connection)
except Exception as e:
    print(f"Error connecting to the database: {e}")

# Access individual DataFrames
author_df = dataframes.get("author")
book_df = dataframes.get("book")
publisher_df = dataframes.get("publisher")
book_language_df = dataframes.get("book_language")
book_author_df = dataframes.get("book_author")
address_status_df = dataframes.get("address_status")
country_df = dataframes.get("country")
address_df = dataframes.get("address")
customer_df = dataframes.get("customer")
customer_address_df = dataframes.get("customer_address")
shipping_method_df = dataframes.get("shipping_method")
cust_order_df = dataframes.get("cust_order")
order_status_df = dataframes.get("order_status")
order_line_df = dataframes.get("order_line")
order_history_df = dataframes.get("order_history")

# Print the first few rows of each DataFrame to verify
if author_df is not None:
    print(author_df.head())
if book_df is not None:
    print(book_df.head())
if publisher_df is not None:
    print(publisher_df.head())
```

```python
#Save all the data frame for further process

author_df.to_csv('author_df.csv', index=False)  # Use index=False to avoid saving the index
book_df.to_csv('book_df.csv', index=False)
book_language_df.to_csv('book_language.csv', index=False)
book_author_df.to_csv('book_author.csv', index=False)
publisher_df.to_csv('publisher.csv', index=False)
shipping_method_df.to_csv('shipping_method.csv', index=False)
address_df.to_csv('address.csv', index=False)
address_status_df.to_csv('address_status.csv', index=False)
customer_address_df.to_csv('customer_address.csv', index=False)
customer_df.to_csv('customer.csv', index=False)
cust_order_df.to_csv('cust_order.csv', index=False)
country_df.to_csv('country.csv', index=False)
order_history_df.to_csv('order_history.csv', index=False)
order_line_df.to_csv('order_line.csv', index=False)
order_status_df.to_csv('order_status.csv', index=False)
```
