# Feauture Engineering




```python
# Rename columns after merging
df_orders = df_orders.rename(columns={
    'status_id_x': 'order_status_id',  # From order_history or order_status
    'status_id_y': 'address_status_id'  # From cust_address
})



# Feature 1: Total Orders per Customer
df_orders['total_orders'] = df_orders.groupby('customer_id')['order_id'].transform('nunique')

# Feature 2: Days Since Last Order
df_orders['last_order_date'] = df_orders.groupby('customer_id')['order_date'].transform('max')
df_orders['days_since_last_order'] = (pd.Timestamp.now() - df_orders['last_order_date']).dt.days

# Feature 3: Average Order Value
df_orders['order_value'] = df_orders.groupby('order_id')['price'].transform('sum')
df_orders['average_order_value'] = df_orders.groupby('customer_id')['order_value'].transform('mean')

# Feature 4: Order Status Counts
df_orders['status_counts'] = df_orders.groupby(['customer_id', 'status_value'])['order_id'].transform('count')

# Feature 5: Address Status Changes

df_orders['address_status_changed'] = df_orders.groupby('customer_id')['address_status_id'].transform('nunique') > 1

# Feature 6: Shipping Method Preferences
df_orders['shipping_method_preference'] = df_orders.groupby('customer_id')['method_name'].transform(lambda x: x.mode()[0])
```

```python
# Define churn

churn_threshold = 90
df_orders['churn'] = (df_orders['days_since_last_order'] > churn_threshold).astype(int)
```

```python
df_orders.to_csv('df_orders.csv',index=False)
```

```python
# Aggregate Customer-Level Features

# Group by customer_id and aggregate features
df_final = df_orders.groupby('customer_id').agg(
    total_orders=('order_id', 'nunique'),
    days_since_last_order=('days_since_last_order', 'max'),
    average_order_value=('average_order_value', 'mean'),
    completed_orders=('status_value', lambda x: (x == 'Delivered').sum()),
    canceled_orders=('status_value', lambda x: (x == 'Cancelled').sum()),
    address_status_changed=('address_status_changed', 'max'),
    shipping_method_preference=('shipping_method_preference', lambda x: x.mode()[0]),
    churn=('churn', 'max')
).reset_index()
```
